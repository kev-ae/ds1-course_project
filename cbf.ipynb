```python
import pandas as pd
df = pd.read_csv('proc.csv')

#Convert artists to list of artists
df['artists'] = df['artists'].astype(str)
df['artists'] = df['artists'].apply(lambda x: x.split(';'))
df.head()
```

```python
def select_columns(df):
    return df[['track_id', 'artists', 'track_name', 'popularity', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'track_genre']]

df = select_columns(df)
df.head()
```

One Hot Encoding

```python
def ohe(df, column, name):
    idf = pd.get_dummies(df[column])
    features = idf.columns
    idf.columns = [name + "|" + str(i) for i in features]
    idf.reset_index(drop = True, inplace = True)
    return idf
```

Encoding Key

```python
key_encoded = ohe(df, 'key', 'key') * 0.25
key_encoded.head()
```

Encoding Mode

```python
mode_encoded = ohe(df, 'mode', 'mode') * 0.25
mode_encoded.head()
```

Encoding Genre

```python
genre_encoded = ohe(df, 'track_genre', 'track_genre') * 0.5
genre_encoded.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

pop = df[['popularity']].reset_index(drop = True)
scaler = MinMaxScaler()
spop = pd.DataFrame(scaler.fit_transform(pop), columns=pop.columns)
spop.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_mat = tfidf.fit_transform(df['artists'].apply(lambda x: " ".join(x)))
art_df = pd.DataFrame(tfidf_mat.toarray())
art_df.columns = ['artist | ' + i for i in tfidf.get_feature_names_out()]
art_df.reset_index(drop = True, inplace = True)
art_df
```

```python
final = pd.concat([art_df, spop, genre_encoded, mode_encoded, key_encoded], axis=1)
final['track_id'] = df['track_id'].values
final.head()
```

```python
playlist_df = pd.read_csv('playlist.csv')
playlist_df = select_columns(playlist_df)
playlist_df['artists'] = playlist_df['artists'].apply(lambda x: x.split(';'))
```

```python
def playlist_features(feature_set, df):
    #Vector
    playlist_feature_set = feature_set[feature_set['track_id'].isin(df['track_id'].values)] #Don't include tracks already in playlist
    playlist_feature_set = playlist_feature_set.drop(columns = 'track_id')
    playlist_vector = playlist_feature_set.sum(axis=0)
    
    #Frame
    comparison_frame = feature_set[~feature_set['track_id'].isin(df['track_id'].values)] #Don't include tracks already in playlist
    
    return playlist_vector, comparison_frame

playlist_vector, comparison_frame = playlist_features(final, playlist_df)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

recommendation = df[df['track_id'].isin(comparison_frame['track_id'].values)]
recommendation['sim'] = cosine_similarity(comparison_frame.drop('track_id', axis=1).values, playlist_vector.values.reshape(1,-1))[:,0]    
```

```python
top_100 = recommendation.sort_values('sim', ascending=False).head(100)
top_100.to_csv('./recommendations.csv')
top_100.head()
```

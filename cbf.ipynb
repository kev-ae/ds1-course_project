# Content-Based Filtering

Imports and making artist column a column of lists

```python
import pandas as pd
df = pd.read_csv('proc.csv')

#Convert artists to list of artists
df['artists'] = df['artists'].astype(str)
df['artists'] = df['artists'].apply(lambda x: x.split(';'))
df.head()
```

Selecting useful columns

```python
df = df[['track_id', 'artists', 'track_name', 'popularity', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'track_genre']]
df.head()
```

One Hot Encoding

```python
def ohe(df, column, name):
    idf = pd.get_dummies(df[column])
    features = idf.columns
    idf.columns = [name + "|" + str(i) for i in features]
    idf.reset_index(drop = True, inplace = True)
    return idf
```

Encoding Key

```python
key_encoded = ohe(df, 'key', 'key') * 0.25
key_encoded.head()
```

Encoding Mode

```python
mode_encoded = ohe(df, 'mode', 'mode') * 0.25
mode_encoded.head()
```

Encoding Genre

```python
genre_encoded = ohe(df, 'track_genre', 'track_genre') * 0.5
genre_encoded.head()
```

Scaling Popularity between 0-1

```python
from sklearn.preprocessing import MinMaxScaler

pop = df[['popularity']].reset_index(drop = True)
scaler = MinMaxScaler()

spop = pd.DataFrame(scaler.fit_transform(pop), columns=pop.columns)
spop.head()
```

Calculating tf-idf score of each artist to determine artist prevalance in a song

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()

#Convert artist list to tfidf-matrix of artists
tfidf_mat = tfidf.fit_transform(df['artists'].apply(lambda x: " ".join(x)))

#Make the tfidf matrix its own dataframe
art_df = pd.DataFrame(tfidf_mat.toarray())
art_df.columns = ['artist | ' + i for i in tfidf.get_feature_names_out()]

#Reset index
art_df.reset_index(drop = True, inplace = True)
art_df.head()
```

Creating final feature dataframe

```python
final = pd.concat([art_df, spop, genre_encoded, mode_encoded, key_encoded], axis=1)
final['track_id'] = df['track_id'].values
final.head()
```

Converting playlist dataframe to its own vector, removing all playlist songs from final feature dataframe

```python
def playlist_features(feature_set, df):
    #Vector
    #add all tracks that are both in the playlist and final df 
    playlist_feature_set = feature_set[feature_set['track_id'].isin(df['track_id'].values)]
    
    #Drop track id column and sum all values in a column, convert to one vector
    playlist_feature_set = playlist_feature_set.drop(columns = 'track_id')
    playlist_vector = playlist_feature_set.sum(axis=0)
    
    #Song dataframe without any songs already in playlist 
    comparison_frame = feature_set[~feature_set['track_id'].isin(df['track_id'].values)] #Don't include tracks already in playlist
    
    return playlist_vector, comparison_frame

playlist_vector, comparison_frame = playlist_features(final, playlist_df)
```

Run cosine similarity algorithm between all dataframe songs and the playlist vector

```python
from sklearn.metrics.pairwise import cosine_similarity
#Trim out all playlist songs from main dataframe 
recommendation = df[df['track_id'].isin(comparison_frame['track_id'].values)] 

#Go through each song in the dataframe and calculate similarity score with playlist vector
recommendation['sim'] = cosine_similarity(comparison_frame.drop('track_id', axis=1).values, playlist_vector.values.reshape(1,-1))[:,0]    
```
